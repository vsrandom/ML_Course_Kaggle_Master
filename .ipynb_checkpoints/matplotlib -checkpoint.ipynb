```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=np.arange(10)
x
sns.distplot(x)
```

```python
y=x**2
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
x=np.linspace(0,20,100)
y=np.linspace(0,30,100)
plt.plot(x,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("My graph")
plt.ylim([0,30])
plt.xlim([0,30])
plt.show()
```

```python
x=np.arange(10)
y=x**2
z=x**3
plt.plot(x,y,'r-',x,z,'b+')
plt.xlabel('X')
plt.ylabel('Y')
plt.title("Two plots")
plt.show()
```

```python
x=np.arange(1,10)
y=np.log(x)
z=x**2
x,y
```

```python
plt.figure(1)
plt.subplot(121)
plt.plot(x,y)
plt.xlabel("X")
plt.ylabel("LoG(x)")
plt.title("Graph1")

plt.subplot(122)
plt.plot(x,z)
plt.xlabel("X")
plt.ylabel("X^2")
plt.title("Graph2")


plt.show()
```

```python
a=x**3
b=np.sin(x)

plt.figure(1)

plt.subplot(121)
plt.plot(x,y)
plt.subplot(122)
plt.plot(x,z)

plt.figure(2)
plt.subplot(121)
plt.plot(x,a)
plt.subplot(122)
plt.plot(x,b)

plt.show()
```
