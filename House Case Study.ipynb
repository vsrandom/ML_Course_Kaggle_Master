Now some of P values are quite bigger and also there is an possiblity of having a corelation between columns like multi colinearity so there is an metric called as VIF (Variance Inflation Factor) which is defined as (1/1-R2).
Now to find VIF what we do is we select one of feature as our label and rest of other features as training data and calculate R2 value. If for any feature R2 value comes high -> VIF value if also high then we can think of removing that column but we also have to make sure that the 'P' value is also high ie greater than 0.05. In genral high VIF value is considered to be something greater than 5.

```python
def vif_cal(input_data,dependent_column):
    vif_dif=pd.DataFrame(columns = ['Var','Vif'])     #dropping the price column
    x_vars=input_data.drop([dependent_column],axis=1)
    xvar_names=x_vars.columns                     #xvar_names contain names of all columns as list
    for i in range(0,xvar_names.shape[0]):      #iterating over all colums and selecting one column at a time
        y=x_vars[xvar_names[i]]
        x=x_vars[xvar_names.drop(xvar_names[i])]  
        rsq=sm.OLS(y,x).fit().rsquared         #getting r square value 
        vif=round(1/(1-rsq),2)
        vif_dif.loc[i]=[xvar_names[i],vif]    #putting column name and vif value for each column
    return vif_dif.sort_values(by='Vif',axis=0,ascending=False,inplace=False)
```
